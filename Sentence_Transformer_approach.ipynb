```python
# from sklearn.model_selection import train_test_split
# X_train, X_val, y_train, y_val = train_test_split(Xtrain, Ytrain, test_size=0.1, random_state=42)


# import pickle
# with open('/content/drive/MyDrive/COLIEE data/data/Xtrain.pkl', 'wb') as f:  # open a text file
#     pickle.dump(X_train, f) # serialize the list


# with open('/content/drive/MyDrive/COLIEE data/data/Xval.pkl', 'wb') as f:  # open a text file
#     pickle.dump(X_val, f) # serialize the list


# with open('/content/drive/MyDrive/COLIEE data/data/Ytrain.pkl', 'wb') as f:  # open a text file
#     pickle.dump(y_train, f) # serialize the list


# with open('/content/drive/MyDrive/COLIEE data/data/Yval.pkl', 'wb') as f:  # open a text file
#     pickle.dump(y_val, f) # serialize the list


import pickle

# open a file, where you stored the pickled data
file = open('/content/drive/MyDrive/COLIEE data/data/Xtrain.pkl', 'rb')
# dump information to that file
X_train = pickle.load(file)

file = open('/content/drive/MyDrive/COLIEE data/data/Ytrain.pkl', 'rb')
# dump information to that file
y_train = pickle.load(file)

# open a file, where you stored the pickled data
file = open('/content/drive/MyDrive/COLIEE data/data/Xval.pkl', 'rb')
# dump information to that file
X_val = pickle.load(file)

file = open('/content/drive/MyDrive/COLIEE data/data/Yval.pkl', 'rb')
# dump information to that file
y_val = pickle.load(file)
```

```python
from sentence_transformers import InputExample
label2int = {"contradiction": 0, "entailment": 1, "neutral": 2}
train_samples = []
for x, y in zip(X_train, y_train):
    if y == 'Y':
        lbl = 'entailment'
    else:
        lbl = 'contradiction'
    train_samples.append(InputExample(texts=[x[0].decode('utf-8'), x[1].decode('utf-8')], label=label2int[lbl]))
```

```python
from sentence_transformers import CrossEncoder
model = CrossEncoder('cross-encoder/nli-deberta-v3-base', num_labels=3)
# model = CrossEncoder('cross-encoder/nli-MiniLM2-L6-H768', num_labels=3)
# model = CrossEncoder('cross-encoder/nli-deberta-v3-xsmall', num_labels=3)
# model = CrossEncoder('cross-encoder/nli-deberta-v3-small', num_labels=3)
```

```python
from sentence_transformers import SentenceTransformer, SentencesDataset, losses
from torch.utils.data import DataLoader
train_dataloader = DataLoader(train_samples, shuffle=True, batch_size=1, collate_fn=model.smart_batching_collate)
```

```python
model.fit(train_dataloader=train_dataloader,
          epochs=5,
          warmup_steps=1000,
          output_path="/content/drive/MyDrive/COLIEE samples/nli_model_H30_EVA/")
```

```python
from sklearn.metrics import classification_report
preds = []
label_mapping = ['contradiction', 'entailment', 'neutral']
for x in X_train:
    scores = model.predict([(x[0].decode('utf-8'), x[1].decode('utf-8'))])
    labels = [label_mapping[score_max] for score_max in scores.argmax(axis=1)]
    if labels[0] == 'entailment':
        preds.append('Y')
    else:
        preds.append('N')
pred_mod = [1 if pred=='Y' else 0 for pred in preds]
grd = [1 if pred=='Y' else 0 for pred in y_train]
print(classification_report(pred_mod, grd))
```

```python
import pickle
with open('/content/drive/MyDrive/COLIEE samples/deberta_base_meta_Xtrain.pkl', 'wb') as f:  # open a text file
    pickle.dump(pred_mod, f) # serialize the list

with open('/content/drive/MyDrive/COLIEE samples/meta_ytrain.pkl', 'wb') as f:  # open a text file
    pickle.dump(grd, f) # serialize the list
```

```python
from sentence_transformers import CrossEncoder
model = CrossEncoder('cross-encoder/nli-roberta-base', num_labels=3)
```

```python
from sentence_transformers import SentenceTransformer, SentencesDataset, losses
from torch.utils.data import DataLoader
train_dataloader = DataLoader(train_samples, shuffle=True, batch_size=1, collate_fn=model.smart_batching_collate)
```

```python
model.fit(train_dataloader=train_dataloader,
          epochs=3,
          warmup_steps=1000,
          output_path="/content/drive/MyDrive/COLIEE samples/nli_model_H30_EVA/")
```

```python
from sklearn.metrics import classification_report
preds = []
label_mapping = ['contradiction', 'entailment', 'neutral']
for x in X_train:
    scores = model.predict([(x[0].decode('utf-8'), x[1].decode('utf-8'))])
    labels = [label_mapping[score_max] for score_max in scores.argmax(axis=1)]
    if labels[0] == 'entailment':
        preds.append('Y')
    else:
        preds.append('N')
pred_mod = [1 if pred=='Y' else 0 for pred in preds]
grd = [1 if pred=='Y' else 0 for pred in y_train]
print(classification_report(pred_mod, grd))
```
